```python
import cv2 as cv
import numpy as np
path=r"C:\Users\asannidhi\Downloads\bnwforopencv.jpg"
img=cv.imread(path)
kernel = np.ones((5,5),np.uint8)
erosion = cv.erode(img,kernel,iterations = 1)
```

```python
from matplotlib import pyplot as plt
```

```python
plt.imshow(erosion)
```

```python
plt.imshow(img)
```

```python
diluted=cv.dilate(img,kernel,iterations=1)
```

```python
plt.imshow(diluted)
```

```python
opened=cv.morphologyEx(img,cv.MORPH_OPEN,kernel)
closed=cv.morphologyEx(img,cv.MORPH_CLOSE,kernel)
```

```python
fig1=plt.imread('opened');
fig2=plt.imread('closed');
```
