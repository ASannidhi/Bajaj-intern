```python
import cv2 as cv
from matplotlib import pyplot as plt
path=r"C:\Users\asannidhi\Downloads\colorforopencv.jpg"
img1=cv.imread(path,cv.IMREAD_GRAYSCALE)
img2=cv.cvtColor(img1, cv.COLOR_RGB2BGR)
ret2,th2 = cv.threshold(img1,0,255,cv.THRESH_BINARY+cv.THRESH_OTSU)
plt.imshow(th2)
```
