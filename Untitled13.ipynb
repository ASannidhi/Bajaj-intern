```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img_path="image_0.png"
img = cv2.imread(img_path)
template=img[250:320,350:430] 
top_left=(350,250)
bottom_right=(430,320)
cv2.rectangle(img,top_left, bottom_right, 255, 2)
imageGray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
templateGray = cv2.cvtColor(template, cv2.COLOR_RGB2GRAY)
result = cv2.matchTemplate(imageGray, templateGray, cv2.TM_CCOEFF_NORMED)
(minVal, maxVal, minLoc, maxLoc) = cv2.minMaxLoc(result)
plt.imshow(img)
```

```python
print(maxVal)
threshold = 0.69
loc = np.where(maxVal>=threshold)
for pt in zip(*loc[::-1]):
    cv2.putText(img, "PRESENT", pt, cv2.FONT_HERSHEY_SIMPLEX, 10, (0,255,0), 5)
plt.imshow(img)
```

```python
threshold = 0.8
loc = np.where(result >= threshold)
for pt in zip(*loc[::-1]):
    cv2.putText(img, "PRESENT", pt, cv2.FONT_HERSHEY_SIMPLEX, 2, (0,255,0), 1)
plt.imshow(img)
```
