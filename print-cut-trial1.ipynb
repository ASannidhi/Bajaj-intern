```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# read image as grayscale
img = cv2.imread('printcut1.png', cv2.IMREAD_GRAYSCALE)

# threshold to binary
thresh = cv2.threshold(img, 100, 255, cv2.THRESH_BINARY)[1]

# apply morphology
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5,5))
morph = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel)

# find contours - write black over all small contours
letter = morph.copy()
cntrs = cv2.findContours(morph, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
cntrs = cntrs[0] if len(cntrs) == 2 else cntrs[1]
for c in cntrs:
    area = cv2.contourArea(c)
    if area < 100:
        cv2.drawContours(letter,[c],0,(0,0,0),-1)

# do canny edge detection
edges = cv2.Canny(letter, 0, 255)

# write results
#cv2.imwrite("K_thresh.png", thresh)
#cv2.imwrite("K_morph.png", morph)
#cv2.imwrite("K_letter.png", letter)
#cv2.imwrite("K_edges.png", edges)

plt.subplot(2, 2, 1)
plt.imshow( thresh)
plt.subplot(2, 2, 2)
plt.imshow( morph)
plt.subplot(2, 2, 3)
plt.imshow( letter)
plt.subplot(2, 2, 4)
plt.imshow( edges)
cv2.waitKey(0)
```
