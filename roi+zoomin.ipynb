```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
img_path="opencv_frame_0.png"
img= cv.imread(img_path)
#plt.imshow(img)

(h,w) = img.shape[:2]
image_gray= cv.cvtColor(img, cv.COLOR_RGB2GRAY)
ROI= np.array([[(250,280),(250,180),(360,180),(360,280)]], dtype= np.int32)

blank= np.zeros_like(image_gray)
region_of_interest= cv.fillPoly(blank, ROI,255)
region_of_interest_image= cv.bitwise_and(image_gray, region_of_interest)

plt.imshow(region_of_interest_image,cmap='gray')
```

```python
def zoom_center(fig, zoom_factor=4):

    y_size = fig.shape[0]
    x_size = fig.shape[1]
    
    # define new boundaries
    x1 = int(0.5*x_size*(1-1/zoom_factor))
    x2 = int(x_size-0.5*x_size*(1-1/zoom_factor))
    y1 = int(0.5*y_size*(1-1/zoom_factor))
    y2 = int(y_size-0.5*y_size*(1-1/zoom_factor))

    # first crop image then scale
    img_cropped = fig[y1:y2,x1:x2]
    return cv.resize(img_cropped, None, fx=zoom_factor, fy=zoom_factor)

# read original 
#img = cv2.imread('original.png')

# call our function
img_zoomed_and_cropped = zoom_center(region_of_interest_image)

# write zoomed and cropped version
#cv.imwrite('zoomed_and_cropped.png', img_zoomed_and_cropped)
#plt.imshow(img_zoomed_and_cropped,cmap='gray')
crop_img = img_zoomed_and_cropped[ 50:350, 180:450]
filename = 'savedImage.png'
cv.imwrite(filename, crop_img)
plt.imshow(crop_img,cmap='gray')
```
